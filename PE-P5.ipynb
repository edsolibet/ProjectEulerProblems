# Smallest Multiple

**Problem 5**: 2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
import numpy as np
import math

dividends = list(range(2, 21)) # list of numbers to get LCM

# Function to check for prime numbers
def is_prime(num):
    sqr = int(np.ceil(np.sqrt(num))) # upper limit to check for prime factors
    if num == 2:
        return 1
    for i in range(2, sqr+1):
        if num % i == 0:
            return 0
    return 1

def prime_factorized(num):
    pr_factors = [] # list of prime factors
    pr_factorized = [] # list of prime factors with counts
    num_cp = np.copy(num) # copy original number to iterate divisions
    for i in range(2, num+1):
        if num_cp % i == 0 and is_prime(i): # check if prime factor; must include number itself if prime
            pr_factors.append(i)
    for j in pr_factors:
        while num_cp % j == 0: # as long as divisible, add prime factor to list then update num
            pr_factorized.append(j)
            num_cp = num_cp/j
    return pr_factorized

# function to count a number within a list
def count_factors(num, arr):
    count = 0
    for i in arr:
        if i == num:
            count+=1
    return count
```

```python
factored = [] # final list of prime factors to be multiplied for LCM
for j in dividends:
    print ("Number {}".format(j))
    fact = prime_factorized(j)
    print (fact)
    for k in fact:
        if k in factored:
            counta = count_factors(k, fact)
            countb = count_factors(k, factored)
            if counta - countb > 0 :
                print ("{} in factored. Appending {} {}x in factored.".format(k, k, counta-countb))
                factored[-1:-1] = [k]*(counta-countb)
        else:
            print ("{} not in factored. Append.".format(k))
            factored.append(k)
        print ("Factored list: {}".format(factored))
        
```

```python
print ("The least common prime factorization is {}".format(factored))
print("The smallest positive number divisible by all of the numbers from 1 to 20 is {}".format(np.prod(factored)))
```
